```python
import pandas as pd
```

```python
stats = pd.read_csv('player_mvp_stats.csv')
del stats['Unnamed: 0']
stats.head()
```

```python
highest_scoring = stats[stats['G'] > 70].sort_values('PTS', ascending=False).head(10)
```

```python
highest_scoring
```

```python
highest_scoring.plot.bar('Player','PTS')
```

```python
highest_scoring_each_season = stats.groupby('Year').apply(lambda x: x.sort_values('PTS', ascending=False).head(1))
```

```python
highest_scoring_each_season
```

```python
highest_scoring_each_season.plot.bar(x=('Player'), y='PTS')
```

```python
stats.corr()['Share'].plot.bar()
```
