```python
import pandas as pd
```

```python
mvps = pd.read_csv('mvps.csv')
mvps.head()
```

```python
mvps = mvps[['Player', 'Pts Won', 'Pts Max', 'Share', 'Year']]
mvps.head()
```

```python
player_stats = pd.read_csv('stats.csv')
player_stats.head()
```

```python
del player_stats['Unnamed: 0']
del player_stats['Rk']
player_stats.head(10)
```

```python
player_stats['Player'] = player_stats['Player'].str.replace('*','',regex=False)
player_stats.head(10)
```

```python
player_stats.groupby(['Player','Year']).get_group(('Courtney Alexander',2001))
```

```python
def convert_to_single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Tm'] == 'TOT']
        row['Tm'] = df.iloc[-1,:]['Tm']
        return row
        
player_stats = player_stats.groupby(['Player','Year']).apply(convert_to_single_row)
```

```python
player_stats
```

```python
player_stats.index = player_stats.index.droplevel()
player_stats.index = player_stats.index.droplevel()
player_stats
```

```python
full_data = player_stats.merge(mvps, how='outer', on=['Player', 'Year'])
full_data.head()
```

```python
full_data[['Pts Won', 'Pts Max', 'Share']]= full_data[['Pts Won', 'Pts Max', 'Share']].fillna(0)
full_data.head()
```

```python
teams = pd.read_csv('team_stats.csv')
teams.head(10)
```

```python
del teams['Unnamed: 0']
teams['Team name'] = teams['Team name'].str.replace('*','',regex=False)
teams.head()
```

```python
nicknames = {}

with open('nicknames.csv', 'r', encoding='utf-8') as f:
    decode = f.read().split('\n')

for _ in decode[1:]:
    abbrev, name = _.split(',')
    nicknames[abbrev] = name
```

```python
full_data['Team name'] = full_data['Tm'].map(nicknames)
full_data.head()
```

```python
full_data = full_data.merge(teams, how='outer', on=('Team name','Year'))
```

```python
full_data
```

```python
full_data = full_data.apply(pd.to_numeric, errors='ignore')
full_data.dtypes
```

```python
full_data['GB'].unique()
```

```python
full_data['GB'] = full_data['GB'].str.replace('—','0',regex=False)
full_data['GB'].unique()
```

```python
full_data['GB'] = pd.to_numeric(full_data['GB'])
full_data.dtypes
```

```python
full_data.to_csv('player_mvp_stats.csv')
```
